```python
# Import iris data and implement k-means clustering algorithm using euclidean distance and calculate accuracy of it over iris data keep it short and simple


import numpy as np
import sklearn as sk
import sklearn.datasets
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
# Load the iris dataset
iris = load_iris()

X = iris.data
y = iris.target
# Implement k-means clustering algorithm
k=3
centroids = X[np.random.choice(X.shape[0], k, replace=False)]

def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))
def k_means(X, centroids):
    while True:
        # Assign labels based on closest centroid
        labels = np.array([np.argmin([euclidean_distance(x, c) for c in centroids]) for x in X])
        new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)])
        if np.all(centroids == new_centroids):
            break
        centroids = new_centroids
    return labels
# Run k-means algorithm
predicted_labels = k_means(X, centroids)
# Calculate accuracy
accuracy = np.mean(predicted_labels == y)
print(f'Accuracy: {accuracy:.2f}')



```
